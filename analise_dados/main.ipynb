```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Carregue o dataset
df = pd.read_csv('train_and_test2.csv')

# Verifique as primeiras linhas do dataset
df.head()
```

```python
# InformaÃ§Ã£o geral sobre o dataset
df.info()

# EstatÃ­sticas descritivas
df.describe()
```

```python
# DistribuiÃ§Ã£o das variÃ¡veis
sns.countplot(x='Survived', data=df)
plt.title('DistribuiÃ§Ã£o de Sobreviventes')
plt.show()

# AnÃ¡lise da relaÃ§Ã£o entre idade e sobrevivÃªncia
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Survived'] == 0]['Age'], bins=30, color='red', label='NÃ£o Sobreviventes')
sns.histplot(df[df['Survived'] == 1]['Age'], bins=30, color='green', label='Sobreviventes')
plt.title('DistribuiÃ§Ã£o de Idade por SobrevivÃªncia')
plt.legend()
plt.show()

# AnÃ¡lise da relaÃ§Ã£o entre classe e sobrevivÃªncia
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Classe x SobrevivÃªncia')
plt.show()

# AnÃ¡lise da relaÃ§Ã£o entre gÃªnero e sobrevivÃªncia
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('GÃªnero x SobrevivÃªncia')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Carregue o dataset
df = pd.read_csv('train_and_test2.csv')

# Verifique as primeiras linhas do dataset
df.head()
```

```python
# Informação geral sobre o dataset
df.info()

# Estatísticas descritivas
df.describe()
```

```python
# Distribuição das variáveis
sns.countplot(x='Survived', data=df)
plt.title('Distribuição de Sobreviventes')
plt.show()

# Análise da relação entre idade e sobrevivência
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Survived'] == 0]['Age'], bins=30, color='red', label='Não Sobreviventes')
sns.histplot(df[df['Survived'] == 1]['Age'], bins=30, color='green', label='Sobreviventes')
plt.title('Distribuição de Idade por Sobrevivência')
plt.legend()
plt.show()

# Análise da relação entre classe e sobrevivência
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Classe x Sobrevivência')
plt.show()

# Análise da relação entre gênero e sobrevivência
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Gênero x Sobrevivência')
plt.show()
```
